```python
# Individual histograms using seaborn for grouping the following continuous variables: age groups, distance from home, years at company
```

```python
df_company = attrition
```

```python
# Attrition by Age Group
```

```python
def Age(a):
    if a <= 25:
        return 1
    elif a > 25 and a <= 32:
        return 2
    elif a > 32 and a <= 40:
        return 3
    elif a > 40 and a <= 50:
        return 4
    else:
        return 5

df_company["Age_group"] = df_company["Age"].apply(lambda x: Age(x))
df_company["Age_group"].value_counts()
sns.countplot(x = "Age_group", hue = "Attrition", data = df_company)
```

```python
# Looking at the plot demonstating the relationship between attrition and age group, 
# you can see that employees in the under 25 age group tend to quit more often, as well as, employees in the 25 and 32 age group.
```

```python
# Attrition by Distance From Home
```

```python
def DistanceFromHome(d):
    if d <= 5:
        return 1
    elif d > 5 and d <= 10:
        return 2
    elif d > 10 and d <= 15:
        return 3
    elif d > 15 and d <= 20:
        return 4
    elif d > 20 and d <= 25:
        return 5
    else:
        return 6
    
df_company["DistanceFromHome_group"] = df_company["DistanceFromHome"].apply(lambda x: DistanceFromHome(x))
df_company["DistanceFromHome_group"].value_counts()
sns.countplot(x = "DistanceFromHome_group", hue="Attrition", data = df_company)
```

```python
# Looking at the plot demonstating the relationship between attrition and distance from home, 
# you can see that employees living a distance of 20-25 miles or over 25 miles from home tend to quit more often.
```

```python
# Attrition by Years at Company
```

```python
def YearsAtCompany(t):
    if t <= 1:
        return 1
    elif t > 1 and t <= 5:
        return 2
    elif t > 5 and t <= 10:
        return 3
    elif t > 10 and t <= 20:
        return 4
    elif t > 20 and t <= 30:
        return 5
    else:
        return 6

df_company["YearsAtCompany"] = df_company["YearsAtCompany"].apply(lambda x:YearsAtCompany(x))
df_company["YearsAtCompany"].value_counts()
sns.countplot(x = "YearsAtCompany", hue = "Attrition", data = df_company)
```

```python
# Looking at the plot demonstrating the relationship between attrition and years at company, 
# you can see that employees working for the company less than a year tend to quit more often, as well as, those having worked between 1 and 5 years.
```

```python
# Function to generate stacked plots between two variables

def stacked_plot(df, group, target):

    fig, ax = plt.subplots(figsize = (6,4))
    temp_df = (df.groupby([group, target]).size()/df.groupby(group)[target].count()).reset_index().pivot(columns=target, index=group, values=0)
    temp_df.plot(kind = 'bar', stacked = True, ax = ax, color = ["green", "darkred"])
    ax.xaxis.set_tick_params(rotation = 0)
    ax.set_xlabel(group)
    ax.set_ylabel('Attrition')
```
